```python
import pandas as pd, numpy as np
import scipy as sp
import scipy.stats
import json

def calcul (test):
    
    test30 = test.iloc[0:30]
   
    listall = test30['all'].values.tolist()
    print(len(listall))

    
    if len(listall)==0:
        return [], []
    
    else:
        i=1
        for item in listall: 
            test30 = test30.replace(item, i)
            i=i+1
             
            
        test30 = test30.where(test30<31)


        listall = test30['all'].values.tolist()
        list10 = test30['10'].values.tolist()
        list20 = test30['20'].values.tolist()
        list30 = test30['30'].values.tolist()
        list40 = test30['40'].values.tolist()
        list50 = test30['50'].values.tolist()


        #number of items which exist in both lists 
        a=len(set(listall)&set(list10))
        b=len(set(listall)&set(list20))
        c=len(set(listall)&set(list30))
        d=len(set(listall)&set(list40))
        e=len(set(listall)&set(list50))
        overlap = [a, b, c, d, e]
        print(overlap)

        #spearman corr w/ nan processing
        spear10 = sp.stats.mstats.spearmanr(listall, list10)
        spear20 = sp.stats.mstats.spearmanr(listall, list20)
        spear30 = sp.stats.mstats.spearmanr(listall, list30)
        spear40 = sp.stats.mstats.spearmanr(listall, list40)
        spear50 = sp.stats.mstats.spearmanr(listall, list50)
        correl = [spear10.correlation, spear20.correlation, spear30.correlation, spear40.correlation, spear50.correlation]
        print(correl)

    return overlap, correl
```

```python
with open ('idJson200.json') as id_json:
    id_array = json.load(id_json)

for box in id_array: 
    if box != 'box':
        id_set = id_array[box][0]
        id_table = pd.DataFrame(id_set, dtype='float64')
        overlap, correl = calcul(id_table)
        id_set['overlap'] = overlap 
        id_set['correl'] = correl
        id_array[box][0] = id_set
with open ('result200.json','w') as result_json:
    json.dump(id_array, result_json) 
```
