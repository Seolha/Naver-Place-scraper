```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd 
import numpy as np
import scipy as sp
import scipy.stats
```

```python
def extract(url,age,lat_min, lng_min, lat_max, lng_max):
    
    r = requests.get(url)
    c = r.content

    html = BeautifulSoup(c, "html.parser")

    body = html.find("body")
    script = body.find_all("script")

    restaurants = script[2].text

    res_array = restaurants[19:]

    res_data = json.loads(res_array)
    if age == '0':
         query_name = '[bounds:'+lat_min+';'+lng_min+';'+lat_max+';'+lng_max+']'+'[filterId:r09680][query:맛집]'
    else:
         query_name = '[age:'+age+']'+'[bounds:'+lat_min+';'+lng_min+';'+lat_max+';'+lng_max+']'+'[filterId:r09680][query:맛집]' #[bound:lat;lng;lat;lng][filterId:][query:]

    item_list = res_data['businesses'][query_name]['items']
    length = len(item_list)
    id_list = list()
    i = 0

    #for items in itemList: 
    if length >= 30:
        while i<30:
            id_list.append(item_list[i]['id'])
            i=i+1

    return id_list

 
```

```python
def queryMaker(lat_min, lng_min, lat_max, lng_max) :

    url_all = "https://store.naver.com/restaurants/list?bounds="+lat_min+"%3B"+lng_min+"%3B"+lat_max+"%3B"+lng_max+"&filterId=r09680&query=%EB%A7%9B%EC%A7%91"
    url10 = "https://store.naver.com/restaurants/list?age=10&bounds="+lat_min+"%3B"+lng_min+"%3B"+lat_max+"%3B"+lng_max+"&filterId=r09680&query=%EB%A7%9B%EC%A7%91"
    url20 = "https://store.naver.com/restaurants/list?age=20&bounds="+lat_min+"%3B"+lng_min+"%3B"+lat_max+"%3B"+lng_max+"&filterId=r09680&query=%EB%A7%9B%EC%A7%91"
    url30 = "https://store.naver.com/restaurants/list?age=30&bounds="+lat_min+"%3B"+lng_min+"%3B"+lat_max+"%3B"+lng_max+"&filterId=r09680&query=%EB%A7%9B%EC%A7%91"
    url40 = "https://store.naver.com/restaurants/list?age=40&bounds="+lat_min+"%3B"+lng_min+"%3B"+lat_max+"%3B"+lng_max+"&filterId=r09680&query=%EB%A7%9B%EC%A7%91"
    url50 = "https://store.naver.com/restaurants/list?age=50&bounds="+lat_min+"%3B"+lng_min+"%3B"+lat_max+"%3B"+lng_max+"&filterId=r09680&query=%EB%A7%9B%EC%A7%91"

    list_all = extract(url_all, '0', lat_min, lng_min, lat_max, lng_max)
    list10 = extract(url10, '10',lat_min, lng_min, lat_max, lng_max)
    list20 = extract(url20, '20',lat_min, lng_min, lat_max, lng_max)
    list30 = extract(url30, '30',lat_min, lng_min, lat_max, lng_max)
    list40 = extract(url40, '40',lat_min, lng_min, lat_max, lng_max)
    list50 = extract(url50, '50',lat_min, lng_min, lat_max, lng_max)

    pack = {'all':list_all, '10':list10, '20':list20, '30':list30, '40':list40, '50':list50 } #dict

    return pack
```

```python
csv_path = "data/grid1.csv"
grid = pd.read_csv(csv_path)
check = 0
idList = {'box':'pack'}

for i in range(200,329):
    lat_min = str(grid.loc[i,'lat_min'])
    lng_min = str(grid.loc[i,'lng_min'])
    lat_max = str(grid.loc[i,'lat_max'])
    lng_max = str(grid.loc[i,'lng_max'])
    print(lat_min)
    newPack = queryMaker(lat_min, lng_min, lat_max, lng_max)
    newKey = 'box'+str(i)
    idList[newKey] = [newPack] 
    check = check +1

print(check) 

   
```
