```python
import pandas as pd 
import pyproj
from pyproj import transform
from pyproj import Proj

csv_path = "gis/grid_1500.csv"
grid = pd.read_csv(csv_path)

grid.head()


    
```

```python
WGS84 = { 'proj':'latlong', 'datum':'WGS84', 'ellps':'WGS84', }

UTMK = {'proj':'tmerc', 'lat_0':'38N', 'lon_0':'127E', 'ellps':'GRS80',
   'x_0':'200000', 'y_0':'500000', 'k':'1',
   'towgs84':'-115.80,474.99,674.11,1.16,-2.31,-1.63,6.43'}

for i in grid.index:
    grid.loc[i,'lat_min'], grid.loc[i,'lng_min']=pyproj.transform(Proj(**UTMK), Proj(**WGS84), grid.loc[i,'xmin'], grid.loc[i,'ymin'])
    grid.loc[i,'lat_max'], grid.loc[i,'lng_max']=pyproj.transform(Proj(**UTMK), Proj(**WGS84), grid.loc[i,'xmax'], grid.loc[i,'ymax'])
    grid.loc[i,'lat_min']=round(grid.loc[i,'lat_min'],6)
    grid.loc[i,'lng_min']=round(grid.loc[i,'lng_min'],6)
    grid.loc[i,'lat_max']=round(grid.loc[i,'lat_max'],6)
    grid.loc[i,'lng_max']=round(grid.loc[i,'lng_max'],6)

grid.head()
    
```
